```python
import matplotlib.pylab as plt
file=open('exponential.txt','r')
```

```python
list=file.readlines()
```

```python
x_dat=[]
y_dat=[]
for i in list:
    x=float(i.split()[0])
    x_dat.append(x)
    y=float(i.split()[1])
    y_dat.append(y)
```

```python
a=10000
plt.grid(True)
plt.ylim([0.8,1.15])
plt.scatter(x_dat[0:a],y_dat[0:a],s=10)
plt.savefig('exponential.pdf')
```
