```python
import numpy as np
import matplotlib.pyplot as plt  
from itertools import product
import random
import sys
```

```python
n=6
comb=product(range(1,7),repeat=n)
x_average=[]
for i in comb:
    x_average.append(np.average(i))
y=[]
x_average_set=set(x_average)   
for i in x_average_set:
    occurrence=x_average.count(i)
    probabilita=occurrence/len(x_average)
    y.append(probabilita*np.sqrt(n))
plt.bar(list(x_average_set), y, width=0.1)
plt.show() 
mean=np.average(range(1,7))
variance=np.mean(np.var(range(1,7)))

dim=100
sample_means=[]
for simulation in range(0,10000):
    sample=[]
    for i in range(1,dim):
        value=random.randint(1,6)   
        sample.append(value)
    sample_means.append(np.average(sample))

sample_means_set=set(sample_means)
y=[]
for i in sample_means_set:
    occurrence=sample_means.count(i)
    probabilita=occurrence/len(sample_means)
    y.append(probabilita*np.sqrt(10))
plt.bar(list(sample_means_set), y, width=0.05)
plt.xlim(0,7)
```

```python
sum=0
sample_mean=[]
for i in range(1,100000):
    value=random.randint(1,6)
    sum+=value
    sample_mean.append(sum/i)

plt.plot(sample_mean)
plt.show()

sample_mean_set=set(sample_mean)   
y=[]
for i in sample_mean_set:
    occurrence=sample_mean.count(i)
    probabilita=occurrence/len(sample_mean)
    y.append(probabilita*np.sqrt(len(sample_mean)))
plt.bar(list(sample_mean_set), y, zorder=1)
plt.show() 
```
